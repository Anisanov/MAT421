# Optimization of ODE Approximations

```python
# Imports
import numpy as np
import matplotlib.pyplot as plt

# Defining ODE approximation function
def ODEapprox(h):
    # Setting function parameters
    x = np.arange(0, 1 + h, h) # Inputs
    f = lambda x, s: np.sin(x) # Outputs
    f0 = 0 # Initial Condition

    # Explicit Euler Method
    s = np.zeros(len(x))
    s[0] = f0
    for i in range(0, len(x) - 1):
        s[i + 1] = s[i] + h * f(x[i], s[i])

    # Calculating Error
    errors = []
    for i in range(len(x)):
        error = abs((1-np.cos(x[i])) - s[i])
        errors.append(error)

    # Displaying Output
    plt.style.use("seaborn")
    plt.figure(figsize = (12, 8))
    plt.plot(x, s, "r--", label="Approximate")
    plt.plot(np.arange(0, 1, 0.001), 1-np.cos(np.arange(0, 1, 0.001)), "b", label="Exact")
    plt.title("ODE Approximate and Exact Solution")
    plt.xlabel("x")
    plt.ylabel("f(x)")
    plt.legend(loc="upper left")
    plt.show()
    print("Errors at each point: ", errors)
```

As the step size, h, increases the approximation tends towards the analytical solution. The following is an approximation of $\frac{d f(x)}{dx} = sin(x)$ with initial condition $f_{0}=0$. The exact solution is $f(x)=1-cos(x)$

```python
ODEapprox(1)
```
