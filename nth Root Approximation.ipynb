# nth Root Approximation

```python
# Imports
import numpy as np

# Defining Function to return nth root approximation
def my_nth_root(x, n, tol):

  # Starting with a random guess
  a = np.random.randint(0, high=100, size=None, dtype=int)

  # Initializing function with its derivative
  f = lambda y: y**n - x
  df = lambda y: n * y ** (n - 1)

  # If guess falls within the tolerance, return guess as approximate solution
  if abs(f(a)) < tol:
    return a  
  # Loop using Newtonian method until f(a) falls within the tolerance
  while abs(f(a)) > tol:
    a = a - f(a)/df(a)
    if abs(f(a)) < tol:
      return a
```

```python
# Outputting approximations
print("Approximation of square root of 9: ", my_nth_root(9, 2, 0.01))
print("Approximation of cube root of 64: ", my_nth_root(64, 3, 0.01))
print("Approximation of 5th root of 32: ", my_nth_root(32, 5, 0.01))
```
