## Taylor Series Approximation

Taylor Series Approximation is a method of approximating continuous functions as an infinite series expressed in terms of the function's derivatives at a particular point. An example of this approximating f(x) = sin(x).

```python
# Imports
import numpy as np
import matplotlib.pyplot as plt

def TaylorSeriesApprox(input):

    # Setting style for plots
    plt.style.use('seaborn')

    # Initializing variables
    x = np.linspace(-4, 4, 8000)
    y = np.zeros(len(x))
    labels = ["First Order", "Third Order", "Fifth Order", "Seventh Order"]
    xvalues = np.full((4,1), input)
    yvalues = []
    error = []

    # Looping over to find sum at 1st, 3rd, 5th and 7th order Taylor Series
    for n, label in zip(range(4), labels):
        y = y + ((-1) ** n * (x) ** (2 * n + 1)) / np.math.factorial(2 * n + 1)

        yvalues.append(y[(input * 1000) - 4000])
        plt.plot(x,y, label = label)

    # Calculating error
    for i in yvalues:
        error.append(abs(np.sin(input) - i))

    # Marking values of point on graph
    plt.scatter(xvalues, yvalues, marker="D", c="red")

    # Displaying analytic graph 
    plt.plot(x, np.sin(x), "k", label = "Analytic") 
    
    # Displaying approximation
    plt.title("Taylor Series Approximation")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend()
    plt.show()

    # Printing Output
    for i in range(len(error)):
        print("First order Taylor Series Approximation: ", yvalues[i], " Error: ", error[i])
```

The above function approximates f(x) = sin(x) at a given x and displays the error. The approximation for x = 3 is shown below with the approximations highlighted with a red diamond.

```python
TaylorSeriesApprox(3)
```
