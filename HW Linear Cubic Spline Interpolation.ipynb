# Linear/Cubic Spline Interpolation

Linear interpolation is defined as a method of estimating a datapoint between two data points  by geometrically rendering a straight line between two adjacent points. This is shown on random points below.

```python
# Imports 
from scipy.interpolate import interp1d
import matplotlib.pyplot
from random import randrange
import matplotlib.pyplot as plt

# Making random points to plot
x = [1, 2, 3, 4, 5]
y = [randrange(10), randrange(10), randrange(10), randrange(10), randrange(10)]

# Linear Interpolation
z = interp1d(x, y)
estimate = z(3.5)

# Plotting Interpolation
plt.figure(figsize = (10,8))
plt.plot(x, y, "-ob")
plt.plot(3.5, estimate, "ro")
plt.title("Linear Interpolation at x = 3.5")
plt.xlabel("x")
plt.ylabel("y")

```

Cubic spline interpolation is a method used to create new points within a set of known points by utilizing a set of piecewise cubic functions. This is shown below by generating random points and utilizing the cubic interpolation function to estimate values in between.

```python
# Imports 
from scipy.interpolate import CubicSpline
import numpy as np

# Making random points to plot
x = [1, 2, 3, 4, 5]
y = [randrange(10), randrange(10), randrange(10), randrange(10), randrange(10)]

# Cubic Interpolation
z = CubicSpline(x, y, bc_type="natural")
input = np.linspace(0, 5, 100)
estimate = z(input)

# Plotting Interpolation
plt.figure(figsize = (10,8))
plt.plot(input, estimate, "r")
plt.plot(x, y, "bo")
plt.title("Cubic Interpolation")
plt.xlabel("x")
plt.ylabel("y")
```
